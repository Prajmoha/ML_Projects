```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
inp=pd.read_csv("Heart - Heart.csv")
```

```python
inp.head()
```

```python
inp=inp.drop(columns='Unnamed: 0')
inp
```

```python
inp['AHD']=inp['AHD'].astype('category')
inp['AHD']=inp['AHD'].cat.codes
inp
```

```python
inp['ChestPain']=inp['ChestPain'].astype('category')
inp['ChestPain']=inp['ChestPain'].cat.codes
inp
```

```python
inp['Thal']=inp['Thal'].astype('category')
inp['Thal']=inp['Thal'].cat.codes
inp
```

```python
inp.isnull().sum()
```

```python
inp.dropna(inplace=True)
```

```python
inp.isnull().sum()
```

```python
inp.shape
```

```python
X=inp.drop(columns='AHD')
X
```

```python
y=inp['AHD']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scale=scaler.fit_transform(X_train)
```

```python
X_test_scale=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train_scale,y_train)
```

```python
y_train_scale_predict=reg.predict(X_train_scale)
y_train_scale_predict
```

```python
reg.score(X_train_scale,y_train)
```

```python
y_test_scale_predict=reg.predict(X_test_scale)
y_test_scale_predict
```

```python
reg.score(X_test_scale,y_test)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm=confusion_matrix(y_test, y_test_scale_predict)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, 
            xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
plt.xlabel("Predict Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
